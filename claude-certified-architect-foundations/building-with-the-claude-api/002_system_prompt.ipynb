```python
# Load environment variables from .env file
from dotenv import load_dotenv

# Populate process environment with values from the local .env file.
load_dotenv()
```

```python
# Create an API client
from anthropic import Anthropic

# display() is injected by IPython at runtime; import it so type checkers resolve it.
from IPython.display import display

# Initialize the Anthropic client (reads API key from environment [ANTHROPIC_BASE_URL, ANTHROPIC_API_KEY]).
client = Anthropic()
# Choose the model used by the helper chat() function.
model = "claude-sonnet-4-6"
```

```python
# Add a user turn to the running conversation list.
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)


# Add an assistant turn to preserve conversation history.
def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


# Send the full message history to Claude and return plain text.
def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message.content[0].text
```

```python
# Make a starting list of messages
messages = []

# Define the system prompt to guide the assistant's behavior.
system = """
You are a patient math tutor.
Do not directly answer a student's question.
Guide them to a solution step by step.
"""

# Add in the initial user question of "How do I solve 5x+3=2 for x?"
add_user_message(messages, "How do I solve 5x+3=2 for x?")

# Pass the list of messages to the 'chat' to get an answer
answer = chat(messages, system=system)

# Show the messages and the final answer
display(messages)
display(answer)
```

```python
# Make a starting list of messages
messages = []

# Add in the initial user question requesting the Python function
add_user_message(
    messages,
    "Write a Python function that takes a string, checks for duplicate characters.",
)

# Pass the messages plus an inline system prompt to 'chat' to get an answer
answer = chat(
    messages, system="You are a Python engineer who writes very concise code."
)

# Show the messages and the final answer
display(messages)
display(answer)
```
