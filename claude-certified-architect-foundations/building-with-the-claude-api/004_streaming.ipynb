```python
# Load environment variables from .env file
from dotenv import load_dotenv

# Populate process environment with values from the local .env file.
load_dotenv()
```

```python
# Create an API client
from anthropic import Anthropic

# display() is injected by IPython at runtime; import it so type checkers resolve it.
from IPython.display import display

# Initialize the Anthropic client (reads API key from environment [ANTHROPIC_BASE_URL, ANTHROPIC_API_KEY]).
client = Anthropic()
# Choose the model used by the helper chat() function.
model = "claude-sonnet-4-6"
```

```python
# Add a user turn to the running conversation list.
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)


# Add an assistant turn to preserve conversation history.
def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


# Send the full message history to Claude and return plain text.
def chat(messages, system=None, temperature=1.0):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature,
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message.content[0].text
```

```python
# Make a starting list of messages
messages: list[dict[str, str]] = []

# Add the user's prompt as the first turn in the conversation.
add_user_message(messages, "Write a 1 sentence description of a fake database")

# Request a streaming response so tokens arrive incrementally instead of all at once.
stream = client.messages.create(
    model=model,
    messages=messages,
    max_tokens=1000,
    stream=True
)

# Iterate over the server-sent events and print each streaming event as it arrives.
for event in stream:
  print(event)

# Show the messages and the final answer
display(messages)
```

```python
messages= []

add_user_message(messages, "Write a 1 sentence description of a fake database")

with client.messages.stream(
    model=model,
    messages=messages,
    max_tokens=1000
) as stream:
    for text in stream.text_stream:
        print(text, end="")

stream.get_final_message()
```
